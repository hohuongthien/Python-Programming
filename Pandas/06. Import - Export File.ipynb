```python
# Import file JSON được lưu ở 1 URL nào đó
import pandas as np

#Khai báo URL là đường dẫn đến file JSON
URL = 'http://raw.githubusercontent.com/BindiChen/machine-learning/master/data-analysis/027-pandas-convert-json/data/simple.json'

df = pd.read_json(URL)

df.head()
```

```python
# Drop row 1 and column 'math'
df.drop(index=[0,2],columns=['math','physics'])
```
